```python
import random
print("*"*35)
print("\t欢迎来到王者峡谷！")
print("*"*35)

role=input("请输入游戏人物:(1.鲁班 2.后羿 3.李白 4.孙尚香 5.貂蝉 6.诸葛亮)\n")

coins=1000

weapon_list=[]


print("欢迎！{1}来到王者峡谷，当前金币是:{0}".format(coins,role))

while True:
    choice = int(input("请选择:\n 1.购买武器\n 2.打仗\n 3.删除武器\n 4.查看武器\n 5.退出游戏\n"))
    
    if choice==1:
        print("欢迎进到武器库:")
        weapons =[["屠龙刀",500],["樱花枪",400],["98K",1000],["手榴弹",800],["碧血剑",700],["鹅毛扇",800]]
        for weapon in weapons:
            print(weapon[0],weapon[1],sep=" ")
            
        weapname= input("请输入要购买的武器名称:")
        if weapname not in weapon_list:
            for weapon in weapons:
                if weapname== weapon[0]:
                    if coins>= weapon[1]:
                        coins-=weapon[1]
                        weapon_list.append(weapon[0])
                        print("{}购买武器：{}成功！".format(role,weapname))
                        break
                    else:
                        print("金币余额不足，赶紧打仗挣钱!")
                        break
            else:
                print("输入武器名称错误!")
        else:
            print("已经拥有此武器!")
                
    elif choice==2:
        print("进入战场.......")
        if  len(weapon_list)>0:
            print("{}拥有的武器如下:".format(role))
            for weapon in weapon_list:
                print(weapon)

            while True:
                weapname= input("请选择:")

                if weapname in weapon_list:

                    ran1=random.randint(1,20)
                    ran2=random.randint(1,20)

                    if ran1>ran2:
                        print("此局游戏：张飞胜!!!")
                    elif ran1<ran2:

                        coins+=200
                        print("此局游戏:{}胜!金币:{}".format(role,coins))
                    else:
                        print("此局游戏为平局，可以再次对战！")

                    break
                else:
                    print("选择的武器不存在，请重新选择")
        else:
            print("还没有购买武器，请赶紧购买武器去吧！")

        
    elif choice==3:
        print("武器太多，很沉，扔几个......")
        
        if len(weapon_list)>0:
            print("{}拥有的武器如下:".format(role))
            for weapon in weapon_list:
                print(weapon)
                
            while True:     
                weapname= input("请选择要删除的武器名称:")
                if weapname in weapon_list:
                    weapon_list.remove(weapname)
                    
                    for weapon in weapons: 
                        if weapname==weapon[0]:
                            coins+=weapon[1]
                            break
                            
                    break
                else:
                    print("武器名称输入有误!")
                
    elif choice==4:
        print("{}拥有的武器如下:".format(role))
        for weapon in weapon_list:
            print(weapon)
        print("总金币:",coins)
            
    elif choice==5:
        answer=input("确定要离开王者荣耀游戏吗?(yes/no):")
        if answer=="yes":
            break
        
    else:
        print("输入错误，请重新选择！")
```
